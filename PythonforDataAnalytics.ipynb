```python
import pandas as pd
import numpy as np
from sklearn import preprocessing



# Convert the strings in the universities DataFrame to numerical values
universities_array = pd.factorize(universities["Unnamed: 1"])[0]

# Reshape the universities array to a 2D array
universities_array = np.reshape(universities_array, (universities.shape[0], 1))

# Scale the universities array
scaled_data = preprocessing.scale(universities_array)

# Convert the scaled data back to a Pandas DataFrame
DF = pd.DataFrame(scaled_data, index=universities.index, columns=universities.columns)

# Print the descriptive statistics of the scaled data
DF.describe()
```

```python
import numpy as np
normal = np.random.normal(size=1000)
normal = pd.DataFrame(normal)
normal.hist(figsize=(8,8),bins=30)
```

```python
skewed = np.random.exponential(scale=2,size=1000)
skewed = pd.DataFrame(skewed)
skewed.hist(figsize=(8,8),bins=30)
```

```python
transform = skewed.apply(np.sqrt)
transform.hist(figsize=(8,8),bins=30)
```

```python
"LOG transformation to remove skewness"
LOGtransform = (skewed+1).apply(np.log)
LOGtransform.hist(figsize=(8,8),bins=30)
```

```python
"imputing with sklearn"
from sklearn.preprocessing import Imputer
#set few null value
normal["column1"]= np.where(normal["column1"]>10,null,normal["column1"])
#we impute
imp=Imputer(missingvalues='NaN',strategy='mean',axis=0)
imputed_cars=imp.fit_transform(normal)
imputed_cars=pd.DataFrame(imputed_cars,index=normal.index,columns=normal.columns)
imputed_cars.head(10)
#Fills null with the mean value by rows
```
